```python
import pandas as pd
import numpy as np
```

```python
df_x = pd.read_csv("./lstm_x.csv")
df_y = pd.read_csv("./lstm_y.csv")
```

```python
df_x
```

```python
df_y
```

```python
columns_test = []
columns_test.append("ID")
for i in range(60):
    num = "v"+str(i+1)
    columns_test.append(num)
    
df = pd.DataFrame(columns=columns_test)
```

```python
for i in range(len(df_x)):
    new_row = []  
    x_row = df_x.iloc[i]
    y_row = df_y.iloc[i]
    new_row.append(x_row[0])
    for i in range(30):
        new_row.append(x_row[i+1])
        new_row.append(y_row[i+1])
    df_length = len(df)
    df.loc[df_length] = new_row
```

```python
df['ID'] = df['ID'].map(round)
```
